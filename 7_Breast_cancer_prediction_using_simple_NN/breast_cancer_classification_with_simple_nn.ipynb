Importing Dependecies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data collection and Processing

```python
#load the data
data = sklearn.datasets.load_breast_cancer()
```

```python
print(data)
```

```python
# loading the data to a data frame
df= pd.DataFrame(data.data, columns = data.feature_names)
```

```python
#adding target column to the dataframe
df['label'] = data.target
```

Separating the features and target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
X.head()
```

```python
Y.head()
```

Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

Standardize the data

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

Building a neural network

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python


# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python

# training the Meural Network

nn_train= model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualzie the accuracy and loss

```python
#accuracy :
plt.plot(nn_train.history['accuracy'])
plt.plot(nn_train.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python

#loss
plt.plot(nn_train.history['loss'])
plt.plot(nn_train.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
